# To check if Time Series is stationary
Using a stationary series is easy for model prediction because its statistical properties—mean, variance, and autocorrelation—remain constant over time, making future behavior predictable based on past data
1.   Mean value is constant or not
2.   Variance is constant or not
3.   Covariance depends only on lag

Stock prices typically follow a random walk meaning they are non stationary and their mean, variance changes over time.

Whereas the return often behave like stationary noise around zero.

These can be visualized using the rolling mean and rolling standard to give a local estimate.

ADF[augumented dickey fuller test] test can also be used ,Hypotheses

H₀ (Null): Series has a unit root (non-stationary)

H₁ (Alternative): Series is stationary

If p-value < 0.05:
→ Reject unit root → stationary

If p-value > 0.05:
→ Fail to reject → non-stationary

For stock prices:
You almost always fail to reject.

Rolling Volatility of Returns can be used to find that the return are stationary on the mean but not on variance.



```python
# TEST - 01
# Rolling mean and Rolling std

window = 30
# A window of 30  date entries has been choosen to find out mean of the trend of
# stock data to specify stationary or not.

for company in eda_data["company_code"].unique():
    subset = eda_data[eda_data["company_code"] == company].copy()
    subset.set_index("date", inplace=True)

    rolling_mean = subset["close"].rolling(window).mean()
    rolling_std = subset["close"].rolling(window).std()

    plt.figure(figsize=(12,5))
    plt.plot(subset["close"], label="Close Price")
    plt.plot(rolling_mean, label="Rolling Mean (30)")
    plt.plot(rolling_std, label="Rolling Std (30)")
    plt.title(f"{company} - Rolling Statistics")
    plt.legend()
    plt.show()
```

```python
# TEST - 02
# Statistical Test (ADF Test)

from statsmodels.tsa.stattools import adfuller

for company in eda_data["company_code"].unique():
    subset = eda_data[eda_data["company_code"] == company]

    result = adfuller(subset["close"])

    print(f"\n{company}")
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
```

```python
# Now to test the return column values and check for their stationarity.
# following the same methods for check rolling variance and ADF test.

for company in eda_data["company_code"].unique():
    subset = eda_data[eda_data["company_code"] == company]

    result = adfuller(subset["daily_return"].dropna())

    print(f"\n{company} (Returns)")
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
```

```python
window = 30

for company in eda_data["company_code"].unique():
    subset = eda_data[eda_data["company_code"] == company].copy()
    subset.set_index("date", inplace=True)

    rolling_std = subset["daily_return"].rolling(window).std()

    plt.figure(figsize=(12,5))
    plt.plot(rolling_std, label="Rolling Std (30)")
    plt.title(f"{company} - Rolling Statistics")
    plt.legend()
    plt.show()
```

Augmented Dickey-Fuller test indicates that raw price series are non-stationary (unit root present). Rolling mean and variance confirm time-varying behavior. The cleaning that we did during the creation of csv file to produce daily returns column, series become stationary in mean, though variance exhibits clustering, suggesting conditional heteroskedasticity typical of financial time series.

# Daily_Returns properties

```python
# To check if the return are fat - tailed
# What does it mean for a series value to be fat tailed, A distribution is fat-tailed if:
# In practical terms:
# Extreme events occur more frequently than Gaussian theory predicts.
# Financial returns almost always exhibit fat tails.

# Statistical Indicator: Kurtosis

# = 0 → Normal

# > 0 → Fat-tailed (leptokurtic)

# < 0 → Thin-tailed

# To compute kurtosis per stock company -:
```

```python
from scipy.stats import kurtosis

fat_tail_stats = {}

for company in eda_data["company_code"].unique():
    returns = eda_data[
        eda_data["company_code"] == company
    ]["daily_return"].dropna()

    k = kurtosis(returns, fisher=False)  # Pearson kurtosis
    fat_tail_stats[company] = k

print(fat_tail_stats)
```

Visual Confirmation (Histogram vs Normal)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

for company in eda_data["company_code"].unique():
    returns = eda_data[
        eda_data["company_code"] == company
    ]["daily_return"].dropna()

    plt.figure(figsize=(8,4))

    # Histogram
    plt.hist(returns, bins=50, density=True, alpha=0.6)

    # Normal curve
    mu = returns.mean()
    sigma = returns.std()
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
    plt.plot(x, norm.pdf(x, mu, sigma))

    plt.title(f"{company} Return Distribution")
    plt.show()
```

From the plots and numerical values we got, we can say that that daily_return are fat distribution i.e fat tailed which was checked using the kurtosis and by histogram and normal plot, a kurtosis value > 3 indicates leptokurtic curve(fat tailed). The finance related distribution are always leptokurtic and involves high risk and implying higher probability of extreme price movements than predicted by normal distribution assumptions.

```python
# Finding correlation between the 4 stocks,
# basic of which could be interpreted from the combined plot that we drew
# at the beginning, like the IT sector companies had relationship of being
# more highup and lowup going with more time required to catch up to other.

# In this section we would be using the formula for correaltion applied on all 4

# ρ(ij) = Corr(Ri,Rj)
# Where Ri and Rj are daily returns of two stocks.
```

```python
# Pivot to wide format
returns_wide = eda_data.pivot(
    index="date",
    columns="company_code",
    values="daily_return"
)

# Drop initial NaNs
returns_wide = returns_wide.dropna()
returns_wide
# This will show us the values of the daily_return of every company as a new
# column

correlation_matrix = returns_wide.corr()
correlation_matrix

# This will create a scoring matrix based on the formula of correlation
# with itself and the next company and create a matrix form the diagonal would be
# 1, as it is compared with itself.
```

Just by looking at the scores we can observe that the previoulsy predicted correlation was right and the correlation is max between INFY and TCS, both IT sectors. 2nd most correlated are SBI and RELIANCE, one involved in banking other in energy sector.

```python
# The plot of correlation would be as follows -:
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))
plt.imshow(correlation_matrix)
plt.colorbar()
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
plt.title("Stock Return Correlation Matrix")
plt.show()
# 0.80 – 1.00 → Very strong

# 0.60 – 0.80 → Strong

# 0.40 – 0.60 → Moderate

# 0.20 – 0.40 → Weak

# <0.20 → Very weak
```

Correlation analysis was performed on daily returns using a Pearson correlation matrix. IT stocks exhibit strong positive correlation, reflecting sectoral linkage. Banking and energy stocks show moderate correlation with the IT sector. Correlation levels suggest partial but limited diversification benefits across these large-cap equities.

# SEASONALITY

```python
# Seasonality basically tells us that, is there any relationship
# between the stocks return value with the season in which it is being
# traded, like for a simpler example is of a rain product that is being sold
# in summer and in winter season the demand and the amt being traded would
# differ in both.
```

```python
# WEEK-DAY-EFFECT
eda_data["weekday"] = eda_data["date"].dt.weekday

# 0 = monday, 1 = tuesday, 2 = wednesday, 3 = thursday, 4 = friday

weekday_mean = (
    eda_data.groupby(["company_code", "weekday"])["daily_return"]
    .mean()
    .unstack()
)

weekday_mean
# unstack is used to stack all the values of return for different companies
# into a similar 5 columns
# each score will tell us mean value how much trading of that particular
# stock has been done in that respective day.
```

```python
weekday_mean.T.plot(kind="bar", figsize=(10,6))
plt.title("Average Return by Weekday")
plt.ylabel("Mean Daily Return")
plt.show()
```

# General Observation: The "Friday Effect"
There is a clear market-wide pattern where Fridays (4) are significantly bearish. Almost all companies (except Reliance) show negative average returns on Fridays, suggesting traders tend to book profits or exit positions before the weekend.

Wednesdays (2) appear to be the strongest day collectively for the group

# Summary of Strategy Implications (Hypothetical)
Mid-Week Momentum: If you are looking for short-term entry, Tuesday/Wednesday appears to be the "sweet spot" where buying pressure is highest across the board.

Weekend Risk: Holding SBIN, INFY, or TCS over the weekend (buying on Friday) has statistically been a losing trade on average, whereas Reliance tends to close the week strong.

```python
# Month-Day-Effect
eda_data["month"] = eda_data["date"].dt.month

# 1 = jan 2 = feb, 3 = march, 4 = april, 5 = may
# 6 = jun, 7  = jul, 8 = aug, 9 = sept, 10 = oct, 11 = nov,  12 = dec

month_mean = (
    eda_data.groupby(["company_code", "month"])["daily_return"]
    .mean()
    .unstack()
)

month_mean
# unstack is used to stack all the values of return for different companies
# into a similar 5 columns
# each score will tell us mean value how much trading of that particular
# stock has been done in that respective month.
```

```python
month_mean.T.plot(kind="bar", figsize=(10,6))
plt.title("Average Return by month")
plt.ylabel("Mean monthly Return")
plt.show()
```
